```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# check the unti tanformation
# the parameters in 'si' directory is in SI unit
# the parameters in 'au/unscaled' directory is in atomic unit with the same 'number' of those in 'si' directory
# the parameters in 'au/scaled' directory is in atomic unit with the same value of those in 'si' directory
```

```python
tmp = np.loadtxt(fname='./si/rhodiag.data')
tt_si = tmp[:,0]
rho11_si = tmp[:,1]
rho22_si = tmp[:,2]
rho33_si = tmp[:,3]
rho44_si = tmp[:,4]
```

```python
tmp = np.loadtxt(fname='./au/unscaled/rhodiag.data')
tt_unscaled_au = tmp[:,0]
rho11_unscaled_au = tmp[:,1]
rho22_unscaled_au = tmp[:,2]
rho33_unscaled_au = tmp[:,3]
rho44_unscaled_au = tmp[:,4]
```

```python
tmp = np.loadtxt(fname='./au/scaled/rhodiag.data')
tt_scaled_au = tmp[:,0]
rho11_scaled_au = tmp[:,1]
rho22_scaled_au = tmp[:,2]
rho33_scaled_au = tmp[:,3]
rho44_scaled_au = tmp[:,4]
```

```python
tmp = np.loadtxt(fname='./si/curr.data')
curr_si = tmp[:,2]
```

```python
tmp = np.loadtxt(fname='./au/unscaled/curr.data')
curr_unscaled_au = tmp[:,2]
```

```python
tmp = np.loadtxt(fname='./au/scaled/curr.data')
curr_scaled_au = tmp[:,2]
```

```python
tmp = np.loadtxt(fname='./si/engy.data')
energy_si = tmp[:,2]
```

```python
tmp = np.loadtxt(fname='./au/unscaled/engy.data')
energy_unscaled_au = tmp[:,2]
```

```python
tmp = np.loadtxt(fname='./au/scaled/engy.data')
energy_scaled_au = tmp[:,2]
```

```python
energy_scale = 2.721138625E+001
time_scale = 2.418884327E-002
curr_scale = 6.623618238E-003
hbar_scale = 0.6582119
print(5.1 / time_scale * hbar_scale)
```

```python
plt.plot(tt_si,rho44_si)
plt.plot(tt_unscaled_au,rho44_unscaled_au,'-.')
plt.scatter(tt_scaled_au / energy_scale ,rho44_scaled_au,color='b',marker="x",s=2)
plt.xlabel('t')
plt.ylabel('rho44')
plt.legend(['si','unsclaed_au','scaled_au'])

plt.show()
```

```python
plt.plot(tt_si,curr_si)
plt.plot(tt_unscaled_au,curr_unscaled_au,'-.')
plt.scatter(tt_scaled_au / energy_scale, curr_scaled_au * energy_scale,color='b',marker="x",s=2)
plt.xlabel('t')
plt.ylabel('I')
plt.legend(['si','unsclaed_au','scaled_au'])

plt.show()
```

```python
plt.plot(tt_si,energy_si)
plt.plot(tt_unscaled_au,energy_unscaled_au,'-.')
plt.scatter(tt_scaled_au / energy_scale, energy_scaled_au * energy_scale,color='b',marker="x",s=2)
plt.xlabel('t')
plt.ylabel('Esys')
plt.legend(['si','unsclaed_au','scaled_au'])

plt.show()
```
