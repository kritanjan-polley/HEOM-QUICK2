```python
import numpy as np
from scipy import integrate

sys_bath_coupling = 0.5  #eV
band_width = 5.0         #eV
band_center = 0.0        #eV
hbar = 0.658211928
T = 0.01                 #eV
chemical_potential = 0.0 #ev


def fdt(omega, t):
    omega = np.array([omega], dtype=np.float128)
    t = np.array([t], dtype=np.float128)
    fun = np.exp(-1.j * omega[0] * t[0]) \
          * (0.5 * sys_bath_coupling * (band_width**2)) / ((omega - band_center)**2 + band_width**2)\
          / (1.0 + np.exp( -1 * (omega[0] - chemical_potential) / T ))
    #print('fdt(omega,t)',omega , t, fun)
    return fun


def func(x, t):
    fun = np.exp(1.j * x * t) \
          * (0.5 * sys_bath_coupling * (band_width**2)) / ((x - band_center)**2 + band_width**2)\
          / (1.0 + np.exp( (x - chemical_potential) / (T) ))
    return fun

def gauss_Legendre_integral(fun, down_limit, upper_limit, interval = 100, t = 0.0 + 0.j):
    sample_x = np.array([-0.96028986,
                         -0.79666648,
                         -0.52553241,
                         -0.18343464,
                          0.18343464,
                          0.52553241,
                          0.79666648,
                          0.96028986 ])
    weight_x = np.array([0.10122854,
                         0.22238103,
                         0.31370665,
                         0.36268378,
                         0.36268378,
                         0.31370665,
                         0.22238103,
                         0.10122854 ])
    result = 0.j
    inner_interval = 1
    points = np.linspace(down_limit, upper_limit, interval + 1, dtype = np.float128)
    for i in range(interval):
        inner_points = np.linspace(points[i], points[i + 1], inner_interval + 1, dtype = np.float128)
        for j in range(inner_interval):
            a = inner_points[j]
            b = inner_points[j + 1]
            tmp = 0
            for k in range(8):
                tmp = tmp + weight_x[k] * fun(0.5 * (b - a) * sample_x[k] * interval + 0.5 * (b - a) * interval, t)
            result = result + tmp * (b - a) * 0.5
    
    return result 



time_sample_for_corr = 20 + 1
tt_for_corr = np.linspace(-1.0, 1.0, time_sample_for_corr, dtype = np.float128)

# compare the approximation results and the exact correlation function in time domain
# exact results
#bath_corr_exact = np.zeros(time_sample_for_corr, dtype = np.complex128)
#for i in range(time_sample_for_corr):
#     bath_corr_exact[i] = gauss_Legendre_integral(fdt, down_limit = -10, upper_limit = 10, interval = 20, t=tt_for_corr[i])
#    bath_corr_exact[i],err = integrate.quad(fdt, -100, 100, args=(tt_for_corr[i]))
#    print('bath_corr_exact[i]', i, bath_corr_exact[i])
    
#print('Done!')
```
